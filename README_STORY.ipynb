Over the last few months, numerous bills have been introduced at the state level that — if passed — could make it harder for millions of Americans to cast their ballots. I took a look into the numbers as well as surveys about voter sentiment and reasons cited for not voting. I'm interested in documenting and describing these voter suppression developments.

SECOND CHART 

The US Census Bureau has been tracking “Reasons People Don’t Vote” over the years in the Voting and Registration Supplement to the Current Population Survey. I looked at data back to 2008 with an idea to track sentiment over time. 

The survey normally tracks 12 reasons in a multiple choice survey including: Inconvenient polling place, registration problems, bad weather, busy, not interested. 


I combined data from six XLS sheets, which was very manual and prone to error. I see with this week’s classwork that pandas would make XLS merge much easier. With coaching from Kelly (thank you!), I learned how to format my fields as dates and numbers so that the information would be read by Datawrapper.
